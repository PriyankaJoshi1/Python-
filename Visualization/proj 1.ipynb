```python
import pandas as pd
import matplotlib.pylab as plt
import seaborn as sns
```

```python
%matplotlib inline
```

```python
autodata=pd.read_csv('auto_data.csv')
```

```python
autodata.head()
```

```python
def origin(n):
    if n==1:
        return 'USA'
    elif n==2:
        return 'Europe'
    else:
        return 'Asia'
```

```python
autodata['origin']=autodata['origin'].apply(origin)
```

```python
autodata.head(10)
```

```python
sns.pairplot(autodata[['mpg','weight','origin']],hue='origin',size=3)
```
