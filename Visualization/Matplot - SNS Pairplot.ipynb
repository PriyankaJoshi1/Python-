```python
#import required libraries
import pandas as pd
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt
```

```python
#To view plot in notebook
%matplotlib inline
```

```python
#load iris data set
iris_dataset = load_iris()
#view dataset using dataset's print function
print iris_dataset.DESCR
```

```python
#define features and target
x = iris_dataset.data
y = iris_dataset.target

iris_data = pd.DataFrame(x,columns=['sepal length','sepal width','petal length','petal width'])
iris_target = pd.DataFrame(y,columns=['species'])
```

```python
#view target top 5 records
iris_target.head()
```

```python
#Write a user defined function for flower
# 0- Setosa, 1-Veriscolour, 2-Verginica
def flower(num):
    if num==0:
        return 'Setosa'
    elif num==1:
        return 'Veriscolour'
    else:
        return 'Virginica'
#use apply function
iris_target['species'] = iris_target['species'].apply(flower)
```

```python
#view top 10 target records
iris_target.head(10)
```

```python
#create a combined iris dataset
iris = pd.concat([iris_data,iris_target],axis=1)
#preview data using head method
iris.head()
```

```python
# draw the pair plot using sns for sepal length, sepal width, petal length, petal width and with hue species, set the size to 3
# note: hue is variable in data set to map plot aspects to different colors
import seaborn as sns
sns.pairplot(iris,hue='species',size=3)
```
