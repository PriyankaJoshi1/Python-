```python
#import required libraries
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

#### Bar Plot

```python
#Create an object for storing programming languages
programming_language = ('Python','Java','Scala','C++','C#')
y_poistion = np.arange(len(programming_language))
```

```python
#rating of the programming languages
language_rating = [10,9,5,8,7]
```

```python
#start plotting bar chart
plt.bar(y_poistion,language_rating,align='center',alpha=0.7)
#set the y axis lable
plt.ylabel('Rating')
#set the bar value
plt.xticks(y_poistion,programming_language)
#set the title
plt.title('Programming Language Rating')
plt.show()
```

#### Error Bar plot

```python
#Create 15 values betwee 0 and 10
x_values = np.linspace(0,10,15)
```

```python
#view x values for x axis
x_values
```

```python
#create y axis object with sin function using numpy for the defined values of x
y_axis = np.sin(x_values)
```

```python
#start plotting the error bar plot
plt.errorbar(x_values,y_axis,xerr=0.2,yerr=0.3)
plt.show()
```
