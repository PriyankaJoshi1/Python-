```python
# load flights data from sns dataset (built in)
flight_data = sns.load_dataset('flights')
```

```python
# view top 5 records
flight_data.head()
```

```python
# use pivot method to re-arrange the dataset
flight_data = flight_data.pivot('month','year','passengers')
```

```python
# view the dataset
flight_data
```

```python
# use heatmap method to generate the heatmap of the flights data
sns.heatmap(flight_data)
```

```python
sns.pairplot(flight_data,size=1)
```

```python
import pandas as pd
```

```python
df_test = pd.read_csv('auto_data_mpg.csv')
```

```python
df_test.head()
```

```python
def origin(num):
    if num == 1:
        return 'America'
    elif num == 2:
        return 'Asia'
    else:
        return 'Europe'
# Apply
df_test['origin'] = df_test['origin'].apply(origin)
```

```python
df_test.head()
```

```python
sns.pairplot(df_test[['mpg','weight','origin']],hue='origin',diag_kind='hist',size=4)
```

```python
new_fig = sns.PairGrid(df_test[['mpg','weight','origin']])
new_fig.map_upper(plt.scatter)
new_fig.map_lower(sns.kdeplot)
new_fig.map_diag(plt.hist,bins=30)
```
